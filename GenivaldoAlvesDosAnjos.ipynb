```python
# Dados de porcentagens de lucro para a Corretora A
porcentagens_corretora_a = [
    [45, 60, 58],
    [62, 55, 70],
    [38, 48, 64],
    [55, 56, 55],
    [54, 59, 48],
    [65, 55, 60]
]

# Dados de porcentagens de lucro para a Corretora B
porcentagens_corretora_b = [
    [57, 55, 58],
    [50, 52, 59],
    [59, 55, 56],
    [61, 52, 53],
    [55, 58, 54],
    [60, 51, 56]
]

# Função para calcular a média e a mediana de uma lista de porcentagens de lucro
def calcular_media_mediana(porcentagens):
    # Inicializar listas para armazenar as médias e medianas
    medias = []
    medianas = []

    # Calcular média e mediana para cada conjunto de porcentagens
    for conjunto in porcentagens:
        # Calcular a média
        media = sum(conjunto) / len(conjunto)
        medias.append(media)

        # Calcular a mediana
        conjunto_ordenado = sorted(conjunto)
        if len(conjunto) % 2 == 0:
            mediana = (conjunto_ordenado[len(conjunto) // 2 - 1] + conjunto_ordenado[len(conjunto) // 2]) / 2
        else:
            mediana = conjunto_ordenado[len(conjunto) // 2]
        medianas.append(mediana)

    return medias, medianas

# Calcular média e mediana para a Corretora A
medias_corretora_a, medianas_corretora_a = calcular_media_mediana(porcentagens_corretora_a)

# Calcular média e mediana para a Corretora B
medias_corretora_b, medianas_corretora_b = calcular_media_mediana(porcentagens_corretora_b)

# Imprimir os resultados
print("Médias para Corretora A:", medias_corretora_a)
print("Medianas para Corretora A:", medianas_corretora_a)

print("Médias para Corretora B:", medias_corretora_b)
print("Medianas para Corretora B:", medianas_corretora_b)
```

```python
import numpy as np
from scipy import stats

corretora_a = [
    [45, 60, 54],
    [62, 55, 70],
    [38, 48, 64],
    [55, 56, 55],
    [54, 59, 48],
    [65, 55, 60]
]

corretora_b = [
    [57, 55, 58],
    [50, 52, 59],
    [59, 55, 56],
    [61, 52, 53],
    [57, 57, 50],
    [55, 58, 54]
]

# Flatten the profit values for each broker
corretora_a_flat = np.array(corretora_a).flatten()
corretora_b_flat = np.array(corretora_b).flatten()

# Calculate mean and standard deviation for each broker
mean_corretora_a = np.mean(corretora_a_flat)
std_corretora_a = np.std(corretora_a_flat)

mean_corretora_b = np.mean(corretora_b_flat)
std_corretora_b = np.std(corretora_b_flat)

# Calculate probabilities

# Probability of profit being equal to 55 in both brokers
prob_eq_55_corretora_a = stats.norm.cdf(55, loc=mean_corretora_a, scale=std_corretora_a)
prob_eq_55_corretora_b = stats.norm.cdf(55, loc=mean_corretora_b, scale=std_corretora_b)
prob_eq_55_both_brokers = prob_eq_55_corretora_a * prob_eq_55_corretora_b

# Probability of profit being greater than 58 in both brokers
prob_gt_58_corretora_a = 1 - stats.norm.cdf(58, loc=mean_corretora_a, scale=std_corretora_a)
prob_gt_58_corretora_b = 1 - stats.norm.cdf(58, loc=mean_corretora_b, scale=std_corretora_b)
prob_gt_58_both_brokers = prob_gt_58_corretora_a * prob_gt_58_corretora_b

# Probability of profit being between 49 and 53 in both brokers
prob_bt_49_53_corretora_a = (stats.norm.cdf(53, loc=mean_corretora_a, scale=std_corretora_a) -
                             stats.norm.cdf(49, loc=mean_corretora_a, scale=std_corretora_a))
prob_bt_49_53_corretora_b = (stats.norm.cdf(53, loc=mean_corretora_b, scale=std_corretora_b) -
                             stats.norm.cdf(49, loc=mean_corretora_b, scale=std_corretora_b))
prob_bt_49_53_both_brokers = prob_bt_49_53_corretora_a * prob_bt_49_53_corretora_b

# Print the probabilities
print(f"Probability of profit being equal to 55 in both brokers: {prob_eq_55_both_brokers:.4f}")
print(f"Probability of profit being greater than 58 in both brokers: {prob_gt_58_both_brokers:.4f}")
print(f"Probability of profit being between 49 and 53 in both brokers: {prob_bt_49_53_both_brokers:.4f}")
```
